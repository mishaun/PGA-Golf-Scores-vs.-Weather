# Clean Data

### Define function to clean numeric values

```python
test_pattern_numeric = '\d+\.?\d*'
re.findall(test_pattern_numeric, '64 °F')
re.findall(test_pattern_numeric, '0.0 °in')

clean_for_numeric_values('7 °mph')
re.findall(test_pattern_numeric, '7 °mph')
```

```python


def clean_for_numeric_values(value):

    pattern_numeric = '\d+\.?\d*'

    try:
        return float(re.findall(pattern_numeric, value)[0])
    except:
        return None

```

### Define Numeric Columns

```python
cols_numeric = ['Temperature',
 'Dew Point',
 'Humidity',
 'Wind Speed',
 'Wind Gust',
 'Pressure',
 'Precip.']

df_hourly_valid_clean = df_hourly_valid.copy()

df_hourly_valid_clean[cols_numeric] = df_hourly_valid_clean[cols_numeric].applymap(clean_for_numeric_values)
```

```python
df_hourly_valid.head()
df_hourly_valid_clean.head()
```

### Rename Columns for Units

```python
df_hourly_valid_clean.rename(columns=
                             {'Precip.': 'Rain (in.)',
                              'Temperature': 'Temperature (°F)',
                              'Wind Speed': 'Wind Speed (mph)',
                              'Wind Gust': 'Wind Gust (mph)'},
                             inplace=True)
```

```python
df_hourly_valid_clean
```

### Clean up Times

```python
# converting time of day to hour number
df_hourly_valid_clean.info()
df_hourly_valid_clean['Hour'] = pd.to_datetime(df_hourly_valid_clean['Time'], format='%I:%M %p').dt.hour
```

```python
# creating a flag for in-play time
def in_play_time(hour):
    if 7 <= hour <= 19:
        return 1
    else:
        return 0
    
df_hourly_valid_clean['Flag Hour In Play'] = df_hourly_valid_clean['Hour'].apply(in_play_time)

df_hourly_valid_clean
```
